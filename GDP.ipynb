### **Matplotlib Data Visualization**

**1. Basic Line plot**

* **Definition:** A line chart used to visualie trends in data over a period.
* **Example:** Internet Users vs Birth Rate.


```python
plt.figure(figsize=(8,5))
plt.plot(df['BirthRate'], df['InternetUsers'], marker='o')
plt.xlabel('Birth Rate')
plt.ylabel('Internet Users')
plt.title('Birth Rate vs Internet Users')
plt.show()
```

**2. Bar plot**

* **Definition:** A bar chart represents categorical data with rectangular bars.
* **Example:** Internet Users by Income Group.


```python
df_grouped= df.groupby('IncomeGroup')['InternetUsers'].mean()

plt.figure(figsize=(6,4))
df_grouped.plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.xlabel("Income Group")
plt.ylabel("Avg. Internet Users")
plt.title("Average Internet Users by Income Group")
plt.show()
```

**3. Histogram**

* **Definition:** represents the freuency distribution of numeric data.

```python
plt.hist(df['BirthRate'], bins=10)
plt.xlabel("Birth Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Birth Rates")
plt.show()
```

```python
plt.hist(df['InternetUsers'], bins=10)
plt.xlabel("InternetUsers")
plt.ylabel("Frequency")
plt.title("Distribution of Internet Users")
plt.show()
```

**4. Scatter plot**

* **Definition:** used to show relationship between two variables.
* **Example:** Birth Rate vs Internet Users.


```python
plt.scatter(df['BirthRate'], df['InternetUsers'], color='red')
plt.xlabel("Birth Rate")
plt.ylabel("Internet Users")
plt.title("Scatter Plot: Birth Rate vs Internet Users")
plt.show()
```

**5. Box plot**

* **Definition:** A box plot shows the distribution of a dataset based on five summary statistics: min, Q1, median, Q3, max.
* **Example:** Distribution of Internet Users.


```python
plt.boxplot(df['InternetUsers'])
plt.ylabel("Internet Users")
plt.title("Boxplot of Internet Users")
plt.show()
```

### **Seaborn Data Visualization**

**1. Scatter plot**

* **Definition:** A plot with enhanced visualization using seaborn.
* **Example:** Birth Rate vs Internet Users with Income Group.


```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='BirthRate', y='InternetUsers', hue='IncomeGroup', data=df)
plt.title("Seaborn Scatter Plot: Birth Rate vs Internet Users")
plt.show()
```

**2. Histogram plot**

* **Definition:** A histogram with KDE support.
* **Example:** Distribution of Birth Rates.


```python
sns.histplot(df['BirthRate'], bins=10, kde=True, color='purple')
plt.title("Histogram of Birth Rates")
plt.show()
```

**3. Seaborn plot**

* **Definition:** A box plot shows the distribution of data across categories.
* **Example:** Internet Users by Income Group.


```python
sns.boxplot(x='IncomeGroup', y='InternetUsers', data=df, palette='Set2')
plt.title('Box Plot: Internet Users by Income Group')
plt.show()
```

**4. Bar plot**

* **Definition:** A bar with statistical aggregation.
* **Example:** Average Internet Users by Income Group.


```python
sns.barplot(x="IncomeGroup", y="InternetUsers", data=df, estimator=sum, palette="coolwarm")
plt.title("Bar Plot: Avg. Internet Users by Income Group")
plt.show()
```

**5. Pair plot**

* **Definition:** A pair plot visualizes pairwise relationships in a dataset.
* **Example:** Visualizing multiple variables.


```python
sns.pairplot(df, hue="IncomeGroup")
plt.show()
```

**6. dist plot**

* **Definition:** used to plot histograms to show the distribution of a variable.
* Previously, sns.distplot() was used, but it is now deprecated in favor of sns.histplot().


```python
#distplot
sns.distplot(df['InternetUsers']) #Univariate analysis
```

**6. displot**

* sns.displot() is an improved figure-level function for distribution plots.
* It allows multiple distributions in a single figure.


```python
#displot
sns.displot(df['InternetUsers']) #Univariate analysis
```

```python
# Using displot
sns.displot(df, x='BirthRate', bins=20, kde=True, color="green")

plt.title("Distribution Plot of Birth Rate")
plt.show()
```

**7. lmplot**

* sns.lmplot() is used for linear regression plots.
* It fits a best-fit line along with scatter points.


```python
#lmplot-linear model plot
sns.lmplot(data=df,x='InternetUsers',y='BirthRate')
```

```python
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)
```

```python
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True)
```

```python
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup')
```
